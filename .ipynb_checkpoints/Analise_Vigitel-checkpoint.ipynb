### Consulta por faixa de classificação de IMC
<font>O <b>IMC</b> é um dado importado do <b>Vigitel</b>, foi calculado usando peso (kg) e altura (m).<br>
Na rotina seguinte os registro são consultados de acordo com a faixa de classificação.<br>
Os registros <b>válidos</b> são os que possuem valor de <b>IMC</b> com valor > 0 e < 100.</font>

```python
# Analisando a evolucao da obesidade

import vigitel
%matplotlib notebook

dic_colunas = {0:'ANO',1:'VALIDOS',2:'ABAIXO',3:'NORMAL',4:'SOBREPESO',5:'OBESIDADE'}
sql_select = '\
SELECT ANO \
     , SUM(VALIDOS)    AS VALIDOS \
     , SUM(ABAIXO)     AS ABAIXO \
     , SUM(NORMAL)     AS NORMAL \
     , SUM(SOBREPESO)  AS SOBREPESO \
     , SUM(OBESIDADE)  AS OBESIDADE \
FROM( \
SELECT ANO \
     , (CASE WHEN (IMC >  0    AND IMC < 100)  THEN 1 ELSE 0 END) AS VALIDOS \
     , (CASE WHEN (IMC >  0    AND IMC < 18.5) THEN 1 ELSE 0 END) AS ABAIXO \
     , (CASE WHEN (IMC >= 18.5 AND IMC < 25)   THEN 1 ELSE 0 END) AS NORMAL \
     , (CASE WHEN (IMC >= 25   AND IMC < 30)   THEN 1 ELSE 0 END) AS SOBREPESO \
     , (CASE WHEN (IMC >= 30   AND IMC < 100)  THEN 1 ELSE 0 END) AS OBESIDADE \
FROM VIGITEL \
) V \
GROUP BY ANO \
ORDER BY ANO'

# exibindo a consulta
dados = vigitel.consulta(sql_select)
df = vigitel.pd.DataFrame(dados).rename(dic_colunas, axis='columns')
df
```

### Calculando a composição da amostra segundo o IMC
<font>É calculada a participação de cada faixa de <b>IMC</b> (apenas registros válidos).<br></font>

```python
# Obesidade - composicao da amostra
# precisa da celula anterior executa

df = vigitel.pd.DataFrame(dados)

# peso por faixa de IMC
anos = df[0]
validos = df[1]
abaixo = df[2]/df[1]
normal = df[3]/df[1]
sobrepeso = df[4]/df[1]
obesidade = df[5]/df[1]

# criando as barras para grafico
vigitel.plt.bar(anos, abaixo, label = "abaixo", color='cyan') 
vigitel.plt.bar(anos, normal, label = "normal", color='green', bottom=abaixo) 
vigitel.plt.bar(anos, sobrepeso, label = "sobrepeso", color='yellow', bottom=abaixo+normal) 
vigitel.plt.bar(anos, obesidade, label = "obesidade", color='red', bottom=abaixo+normal+sobrepeso)

# titulos no grafico
vigitel.plt.xlabel('anos') 
vigitel.plt.ylabel('percentual') 
vigitel.plt.title('Obesidade (por faixa de IMC)\n evolução anual da composição da amostra') 
  
# exibir legenda 
vigitel.plt.legend()
  
# exibir o grafico 
vigitel.plt.show()
```

```python
# Obesidade - composicao da amostra
# Comparacao entre o primeiro ano e o ultimo ano (2009x2019)
# precisa das celulas anteriores executadas

df = vigitel.pd.DataFrame(dados).transpose()

# rotina para exibir a composicao do IMC
def ExibeIMC (serie):
    fig = vigitel.plt.figure()
    ax = fig.add_axes([0,0,1,1])
    ax.axis('equal')
    labels = ['abaixo', 'normal', 'sobrepeso', 'obesidade']
    colors = ['cyan', 'green', 'yellow', 'red'] 
    sizes = [serie[2], serie[3], serie[4], serie[5]]
    explode = [0,0,0,0.1]
    ax.pie(sizes, labels = labels, colors=colors, explode=explode,  
           autopct='%1.2f%%', shadow=True, startangle=90)
    vigitel.plt.show()

ExibeIMC(df[0])    #ano de 2009
ExibeIMC(df[10])   #ano de 2019 
```

### <hr style="height:2px; border:none; color:#000; background-color:#000; margin-top: 0px; margin-bottom: 0px;"/>

```python
# Analisando a evolucao da obesidade

import vigitel
%matplotlib notebook

dic_colunas = {0:'ANO',\
               1:'M_ABAIXO',2:'M_NORMAL',3:'M_SOBREPESO',4:'M_OBESIDADE',\
               5:'F_ABAIXO',6:'F_NORMAL',7:'F_SOBREPESO',8:'F_OBESIDADE'}
sql_select = '\
SELECT ANO \
     , SUM(M_ABAIXO)     AS M_ABAIXO \
     , SUM(M_NORMAL)     AS M_NORMAL \
     , SUM(M_SOBREPESO)  AS M_SOBREPESO \
     , SUM(M_OBESIDADE)  AS M_OBESIDADE \
     , SUM(F_ABAIXO)     AS F_ABAIXO \
     , SUM(F_NORMAL)     AS F_NORMAL \
     , SUM(F_SOBREPESO)  AS F_SOBREPESO \
     , SUM(F_OBESIDADE)  AS F_OBESIDADE \
FROM('
sql_select = sql_select+'\
SELECT ANO \
     , (CASE WHEN (IMC >  0 AND IMC < 18.5  AND SEXO=1) THEN 1 ELSE 0 END) AS M_ABAIXO \
     , (CASE WHEN (IMC >= 18.5 AND IMC < 25 AND SEXO=1) THEN 1 ELSE 0 END) AS M_NORMAL \
     , (CASE WHEN (IMC >= 25 AND IMC < 30   AND SEXO=1) THEN 1 ELSE 0 END) AS M_SOBREPESO \
     , (CASE WHEN (IMC >= 30 AND IMC < 100  AND SEXO=1) THEN 1 ELSE 0 END) AS M_OBESIDADE'
sql_select = sql_select+'\
     , (CASE WHEN (IMC >  0 AND IMC < 18.5  AND SEXO=2) THEN 1 ELSE 0 END) AS F_ABAIXO \
     , (CASE WHEN (IMC >= 18.5 AND IMC < 25 AND SEXO=2) THEN 1 ELSE 0 END) AS F_NORMAL \
     , (CASE WHEN (IMC >= 25 AND IMC < 30   AND SEXO=2) THEN 1 ELSE 0 END) AS F_SOBREPESO \
     , (CASE WHEN (IMC >= 30 AND IMC < 100  AND SEXO=2) THEN 1 ELSE 0 END) AS F_OBESIDADE '
sql_select = sql_select+'\
FROM VIGITEL) V \
GROUP BY ANO'

# exibindo a consulta
dados = vigitel.consulta(sql_select)
df = vigitel.pd.DataFrame(dados).rename(dic_colunas, axis='columns')
df
```

```python
# Obesidade - composicao da amostra
# precisa da celula anterior executa

df = vigitel.pd.DataFrame(dados)

# peso por faixa de IMC
anos = df[0]
validos = df[1]+df[2]+df[3]+df[4]+df[5]+df[6]+df[7]+df[8]
Mabaixo = df[1]/validos
Mnormal = df[2]/validos
Msobrepeso = df[3]/validos
Mobesidade = df[4]/validos
Fabaixo = df[5]/validos
Fnormal = df[6]/validos
Fsobrepeso = df[7]/validos
Fobesidade = df[8]/validos

def ExibeEvolucaoIMC (anos,abaixo,normal,sobrepeso,obesidade,nomesexo):
    fig = vigitel.plt.figure()
    ax = vigitel.plt.axes()
    ax.plot(anos, abaixo, label = "abaixo", color='cyan', linewidth=3)
    ax.plot(anos, normal, label = "normal", color='green', linewidth=3)
    ax.plot(anos, sobrepeso, label = "sobrepeso", color='yellow', linewidth=3)
    ax.plot(anos, obesidade, label = "obesidade", color='red', linewidth=3)
    # titulos no grafico
    vigitel.plt.xlabel('evolução(anos)') 
    vigitel.plt.ylabel('percentual') 
    vigitel.plt.title('Obesidade\n evolução da composição da amostra '+nomesexo) 
    # exibir legenda 
    vigitel.plt.legend()
    vigitel.plt.show()

# exibindo os graficos
ExibeEvolucaoIMC(anos,Mabaixo,Mnormal,Msobrepeso,Mobesidade,'masculina')
ExibeEvolucaoIMC(anos,Fabaixo,Fnormal,Fsobrepeso,Fobesidade,'feminina')
```

```python
# obesidade por ano e cidade
# peso por faixa de IMC, em relacao aos IMCs validos

import vigitel
%matplotlib notebook

sql_select = '\
SELECT ANO \
     , CIDADE \
     , SUM(VALIDOS)    AS VALIDOS \
     , SUM(OBESIDADE)  AS OBESIDADE \
FROM( \
SELECT ANO \
     , CIDADE \
     , (CASE WHEN (IMC >  0    AND IMC < 100)  THEN 1 ELSE 0 END) AS VALIDOS \
     , (CASE WHEN (IMC >= 30   AND IMC < 100)  THEN 1 ELSE 0 END) AS OBESIDADE \
FROM VIGITEL \
) V \
GROUP BY ANO, CIDADE \
ORDER BY ANO, CIDADE'

# exibindo a evolucao anual do peso da obesidade nas cidades 
dados = vigitel.consulta(sql_select)
df = vigitel.pd.DataFrame(dados)
evolucao_obesidade = {'ano':df[0],'cidade':df[1],'obesidade':df[3]/df[2]}
df = vigitel.pd.DataFrame(evolucao_obesidade).\
     pivot(index='cidade', columns='ano', values='obesidade').rename(vigitel.dic_cidade, axis='index')
df
```

### Analisando hipertensão, diabetes e obesidade
<font>É calculada a participação de cada doença em relação ao total válido.<br>
Os registros <b>válidos</b> são os que possuem valor de <b>IMC</b> com valor > 0 e < 100,<br>
e responderam se possuem, ou não, <b>hipertensão</b> e <b>diabetes</b>.</font>

```python
# Evolução anual de doenças

import vigitel
%matplotlib notebook

sql_select = '\
SELECT ANO \
     , SUM(VALIDOS)      AS VALIDOS \
     , SUM(PRESSAO_ALTA) AS PRESSAO_ALTA \
     , SUM(DIABETES)     AS DIABETES \
     , SUM(SOBREPESO)    AS SOBREPESO \
     , SUM(OBESIDADE)    AS OBESIDADE \
FROM( \
SELECT ANO \
     , (CASE WHEN (PRESSAO_ALTA = 1)        THEN 1 ELSE 0 END) AS PRESSAO_ALTA \
     , (CASE WHEN (DIABETES = 1)            THEN 1 ELSE 0 END) AS DIABETES \
     , (CASE WHEN (IMC >  0  AND IMC < 100) THEN 1 ELSE 0 END) AS VALIDOS \
     , (CASE WHEN (IMC >= 25 AND IMC < 30)  THEN 1 ELSE 0 END) AS SOBREPESO \
     , (CASE WHEN (IMC >= 30 AND IMC < 100) THEN 1 ELSE 0 END) AS OBESIDADE \
FROM VIGITEL \
WHERE PRESSAO_ALTA IN (1,2) AND DIABETES IN (1,2) AND IMC > 0 AND IMC < 100 \
) V \
GROUP BY ANO \
ORDER BY ANO'

# peso por faixa de IMC, em relacao aos IMCs validos
dados = vigitel.consulta(sql_select)
df = vigitel.pd.DataFrame(dados)
evolucao_obesidade = {'ano':df[0],'pressao_alta':df[2]/df[1],\
                      'diabetes':df[3]/df[1],'sobrepeso':df[4]/df[1],'obesidade':df[5]/df[1]}
df = vigitel.pd.DataFrame(evolucao_obesidade)
df
```

### Relação da Obesidade e Hipertensão (2019)
<font>São feitas duas consultas: <b>COM</b> e <b>SEM</b> hipertensão.</font>

```python
# Pressão alta - composicao da amostra

import vigitel
%matplotlib notebook

def MostraPressaoAlta (pressao):
    sql_select = '\
    SELECT SUM(NAO_OBESO) AS NAO_OBESO \
         , SUM(SOBREPESO) AS SOBREPESO \
         , SUM(OBESIDADE) AS OBESIDADE \
    FROM( \
    SELECT (CASE WHEN (IMC >  0  AND IMC < 25)  THEN 1 ELSE 0 END) AS NAO_OBESO \
         , (CASE WHEN (IMC >= 25 AND IMC < 30)  THEN 1 ELSE 0 END) AS SOBREPESO \
         , (CASE WHEN (IMC >= 30 AND IMC < 100) THEN 1 ELSE 0 END) AS OBESIDADE \
    FROM VIGITEL \
    WHERE IMC > 0 AND IMC < 100 AND ANO = 2019 AND PRESSAO_ALTA = '+pressao+') V'
    # resultado da consulta
    dados = vigitel.consulta(sql_select)
    df = vigitel.pd.DataFrame(dados)
    fig = vigitel.plt.figure()
    ax = fig.add_axes([0,0,1,1])
    ax.axis('equal')
    labels = ['nao obeso', 'sobrepeso', 'obesidade']
    colors = ['green', 'yellow', 'red'] 
    sizes = [df[0][0], df[1][0], df[2][0]]
    explode = [0,0,0.1]
    ax.pie(sizes, labels = labels, colors=colors, explode=explode,  
           autopct='%1.2f%%', shadow=True, startangle=90)
    vigitel.plt.show()

# exibindo os graficos
MostraPressaoAlta (pressao='1')     #Com pressão alta
MostraPressaoAlta (pressao='2')     #Sem pressão alta
```

### Relação da Obesidade e Diabetes (2019)
<font>São feitas duas consultas: <b>COM</b> e <b>SEM</b> diabetes.</font>

```python
# Diabetes - composicao da amostra

def MostraDiabetes (diabetes):
    sql_select = '\
    SELECT SUM(NAO_OBESO) AS NAO_OBESO \
         , SUM(SOBREPESO) AS SOBREPESO \
         , SUM(OBESIDADE) AS OBESIDADE \
    FROM( \
    SELECT (CASE WHEN (IMC >  0  AND IMC < 25)  THEN 1 ELSE 0 END) AS NAO_OBESO \
         , (CASE WHEN (IMC >= 25 AND IMC < 30)  THEN 1 ELSE 0 END) AS SOBREPESO \
         , (CASE WHEN (IMC >= 30 AND IMC < 100) THEN 1 ELSE 0 END) AS OBESIDADE \
    FROM VIGITEL \
    WHERE IMC > 0 AND IMC < 100 AND ANO = 2019 AND DIABETES = '+diabetes+') V'
    # resultado da consulta
    dados = vigitel.consulta(sql_select)
    df = vigitel.pd.DataFrame(dados)
    fig = vigitel.plt.figure()
    ax = fig.add_axes([0,0,1,1])
    ax.axis('equal')
    labels = ['nao obeso', 'sobrepeso', 'obesidade']
    colors = ['green', 'yellow', 'red'] 
    sizes = [df[0][0], df[1][0], df[2][0]]
    explode = [0,0,0.1]
    ax.pie(sizes, labels = labels, colors=colors, explode=explode,  
           autopct='%1.2f%%', shadow=True, startangle=90)
    vigitel.plt.show()

# exibindo os graficos
MostraDiabetes (diabetes='1')     #Com diabetes
MostraDiabetes (diabetes='2')     #Sem diabetes
```
